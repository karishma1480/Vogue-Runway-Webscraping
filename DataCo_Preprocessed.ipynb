```python
#remove rows that have missing values
df = df.dropna(axis='index', how='all') 
```

```python
#dropping irrelevant columns
drop_columns = ['Order Zipcode',
                'Product Description',
                'Benefit per order',
                'Customer City',
                'Customer Country',
                'Customer Id',
                'Customer Email',                     
                'Customer Fname',                                                  
                'Customer Lname',                        
                'Customer Password',
                'Customer Street',
                'Customer Zipcode',
                'Department Id',
                'Order Item Cardprod Id',
                'Product Card Id',
                'Product Category Id',
                'Product Status',
                'Product Image',
                'Latitude',
                'Longitude']

df.drop(columns=drop_columns, inplace=True, axis=1)


#Explanation:
#inplace=True: Modifies the DataFrame directly and does not return a new DataFrame.
#df = df.drop(...): Without inplace=True, this returns a new DataFrame with the specified columns dropped
```

```python
#rename columns for clarity
df.rename(columns={'Type': 'Payment Type', 
                        'order date (DateOrders)': 'Order Date', 
                        'shipping date (DateOrders)': 'Shipping Date'},inplace=True)
```

```python
#Standardize date formats to a more readable format
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Shipping Date'] = pd.to_datetime(df['Shipping Date'])

# Extract date part
df['Order Day'] = df['Order Date'].dt.day
df['Shipping Day'] = df['Shipping Date'].dt.day

# Extract month in string format
df['Order Month'] = df['Order Date'].dt.strftime('%B')
df['Shipping Month'] = df['Shipping Date'].dt.strftime('%B')

# Extract year
df['Order Year'] = df['Order Date'].dt.year
df['Shipping Year'] = df['Shipping Date'].dt.year
```

```python
#feature engineering on shipping time

#Calculate Actual Shipping time
df['Actual Shipping Time'] = df['Days for shipping (real)'] - df['Days for shipment (scheduled)']
```

```python
#showing only the delayed times as negative occurences in the Actual shipping time means the shipment is early
df['Shipping Delay'] = df['Actual Shipping Time'].apply(lambda x: x if x > 0 else 0)
```
